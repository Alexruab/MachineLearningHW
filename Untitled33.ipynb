# Удовлетворенность пассажиров авиакомпаний


Gender: Пол пассажиров (женщина, мужчина). 

Customer Type: тип клиента (лояльный клиент, нелояльный клиент). 

Age: Фактический возраст пассажиров.

Type of Travel: Цель полета пассажиров (личная поездка, деловая поездка).

Class: Класс путешествия в самолете пассажиров (Бизнес, Эко, Эко Плюс) 

Flight distance: Расстояние полета в этом путешествии.

Inflight wifi service: Уровень удовлетворенности службой Wi-Fi на борту (0: Неприменимо; 1–5).

Departure/Arrival time convenient: Уровень удовлетворенности удобным временем отправления/прибытия. 

Ease of Online booking: уровень удовлетворенности онлайн-бронированием.

Gate location: Уровень удовлетворенности расположением гейта. 

Food and drink: Уровень удовлетворенности едой и напитками.

Online boarding: уровень удовлетворенности онлайн-посадкой. 

Seat comfort: Уровень удовлетворенности комфортом сиденья.

Inflight entertainment: Уровень удовлетворенности развлечениями на борту. 

On-board service: Уровень удовлетворенности бортовым обслуживанием. 

Leg room service: Уровень удовлетворенности местом для ног. 

Baggage handling: Уровень удовлетворенности обработкой багажа 

Check-in service: Уровень удовлетворенности службой регистрации 

Inflight service:  Уровень удовлетворенности обслуживанием во время полёта.

Cleanliness: Уровень удовлетворенности чистотой. 

Departure Delay in Minutes: Задержка в минутах при вылете.

Arrival Delay in Minutes: Задержка в минутах по прибытии. 

Satisfaction:  уровень удовлетворенности авиакомпании (Satisfaction-удовлетворение,neutral- нейтральность или dissatisfaction-неудовлетворенность).

### Выбросы

### Удалим все строки таблицы, в которых 'Age',  'Flight Distance',  'Departure Delay in Minutes','Arrival Delay in Minutes' выходят за пределы квантилей 0.001, 0.999


```python
data_raw_test.quantile([0.001,.01,.05,.1,.5,.9,.95,.99,.999], numeric_only=True)


```

```python
rows_to_drop = data_raw_test[
    (data_raw_test['Age'] < data_raw_test['Age'].quantile(0.001)) | (data_raw_test['Age'] > data_raw_test['Age'].quantile(0.999)) |
    (data_raw_test['Flight Distance'] < data_raw_test['Flight Distance'].quantile(0.001)) | (data_raw_test['Flight Distance'] > data_raw_test['Flight Distance'].quantile(0.999)) |
    (data_raw_test['Departure Delay in Minutes'] > data_raw_test['Departure Delay in Minutes'].quantile(0.999))].index
data_raw_test = data_raw_test.drop(rows_to_drop)

data_raw_test.describe()

```

```python
rows_to_drop = data_raw_train[
    (data_raw_train['Age'] < data_raw_train['Age'].quantile(0.001)) | (data_raw_train['Age'] > data_raw_train['Age'].quantile(0.999)) |
    (data_raw_train['Flight Distance'] < data_raw_train['Flight Distance'].quantile(0.001)) | (data_raw_train['Flight Distance'] > data_raw_train['Flight Distance'].quantile(0.999)) |
    (data_raw_train['Departure Delay in Minutes'] > data_raw_train['Departure Delay in Minutes'].quantile(0.999))].index

data_raw_train = data_raw_train.drop(rows_to_drop)
data_raw_train.describe()
```

### Бинаризация номинальных признаков и нормализация количественных признаков.

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop='if_binary', sparse=False)
enc.fit(data_raw_train[['Gender', 'Customer Type', 'Type of Travel','Class','satisfaction']])

dummies = pd.DataFrame(enc.transform(data_raw_train[['Gender', 'Customer Type', 'Type of Travel','Class','satisfaction']]),
                       columns=enc.get_feature_names_out(), index=data_raw_train.index)
dummies.head()


data_numerical = data_raw_train[numerical]

data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)

data_raw_train = pd.concat((data_numerical, dummies), axis=1)   
```

```python
data_raw_train.head()
```

```python
enc = OneHotEncoder(drop='if_binary', sparse=False)
enc.fit(data_raw_test[['Gender', 'Customer Type', 'Type of Travel','Class','satisfaction']])

dummies = pd.DataFrame(enc.transform(data_raw_test[['Gender', 'Customer Type', 'Type of Travel','Class','satisfaction']]),
                       columns=enc.get_feature_names_out(), index=data_raw_test.index)
dummies.head()


data_numerical = data_raw_test[numerical]

data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)

data_raw_test = pd.concat((data_numerical, dummies), axis=1)   

```

```python
x_train = data_raw_train.drop('satisfaction_satisfied', axis = 1) 
y_train = data_raw_train['satisfaction_satisfied']

x_train = x_train.to_numpy()
y_train = y_train.to_numpy()

x_test = data_raw_test.drop('satisfaction_satisfied', axis = 1) 
y_test = data_raw_test['satisfaction_satisfied']

x_test = x_test.to_numpy()
y_test = y_test.to_numpy()
```

### Метод kNN


```python
print(type(x_train))
print(type(y_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)

knn.fit(x_train, y_train)
```

```python
y_test_predict = knn.predict(x_test)
err_test  = np.mean(y_test  != y_test_predict)
err_test
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

### Подбор оптимальных параметров


```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(x_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors)

knn.fit(x_train, y_train)

y_test_predict = knn.predict(x_test)
err_test  = np.mean(y_test  != y_test_predict)
print(err_test)

print(confusion_matrix(y_test, y_test_predict))
```

### Cлучайный лес

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 100)
rf.fit(x_train, y_train)

y_test_predict = rf.predict(x_test)

err_test  = np.mean(y_test  != rf.predict(x_test))

print(err_test)
```

```python
print(confusion_matrix(y_test, y_test_predict))
```

### Extremely Randomized Trees

```python
ert = ensemble.ExtraTreesClassifier(n_estimators = 100).fit(x_train, y_train)

y_test_predict = ert.predict(x_test)

err_test  = np.mean(y_test  != ert.predict(x_test))

print(err_test)
print(confusion_matrix(y_test, y_test_predict))
```

### AdaBoost

```python
ada = ensemble.AdaBoostClassifier(n_estimators = 1000)
ada.fit(x_train, y_train)

y_test_predict = ada.predict(x_test)
err_test = np.mean(y_test != ada.predict(x_test))

print(err_test)
print(confusion_matrix(y_test, y_test_predict))
```

### GBT

```python
gbt = ensemble.GradientBoostingClassifier(n_estimators = 1000)
gbt.fit(x_train, y_train)

y_test_predict = gbt.predict(x_test)
err_test = np.mean(y_test != gbt.predict(x_test))

print(err_test)
print(confusion_matrix(y_test, y_test_predict))
```

### Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
lgr.fit(x_train, y_train)

y_test_predict = lgr.predict(x_test)
err_test = np.mean(y_test != lgr.predict(x_test))

print(err_test)
print(confusion_matrix(y_test, y_test_predict))
```

### Нейронная сеть (полносвязная, 2 скрытых слоя)

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='adam', activation= 'logistic' ,alpha=1e-5, hidden_layer_sizes=(46, 12,), random_state=1, max_iter= 500)

clf.fit(x_train, y_train)

y_test_predict = clf.predict(x_test)
err_test  = np.mean(y_test  != y_test_predict)
err_test
```

```python
print(confusion_matrix(y_test, y_test_predict))
```

### Из всех методов лучший результат получен при использовании нейронной сети и Randomized Trees.

```python
clf = MLPClassifier(solver='adam', activation= 'logistic' ,alpha=1e-6, hidden_layer_sizes=(46, 12,), random_state=1, max_iter= 500)
clf.fit(x_train, y_train)

y_test_predict = clf.predict(x_test)
err_test  = np.mean(y_test  != y_test_predict)
err_test
```

```python
Alpha = [1e-2, 1e-4, 1e-6, 1e-8, 1e-10]

err_test = []

err_train = []


for Alpha in Alpha:
    clf = MLPClassifier(solver='adam', activation= 'logistic' ,alpha=Alpha, hidden_layer_sizes=(46, 12,), random_state=1, max_iter= 500)
    clf.fit(x_train, y_train)

    y_train_predict = clf.predict(x_train)
    y_test_predict = clf.predict(x_test)
    err_train.append(np.mean(y_train != y_train_predict))
#    train_acc.append(accuracy_score(y_train, y_train_pred))
    err_test.append(np.mean(y_test != y_test_predict))
#    test_acc.append(accuracy_score(y_test, y_test_pred))
```

```python
Alpha = ["1e-2", "1e-4", "1e-6", "1e-8", "1e-10"]

print("Минимальная ошибка на тестовой выборки:", min(err_test))
print("Минимальная ошибка на тестовой выборки:", min(err_train))

plt.plot(Alpha, err_train, 'g-o',label='err_test')
plt.plot(Alpha, err_test, 'b-o',label='err_train')
plt.legend()
plt.show()
```
